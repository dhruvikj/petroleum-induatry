```python
pg = float(input("Give Pressure(KG/cm2) :- "))
Tg = float(input("Give Temperature(celcius) :- "))
s = float(input("Give specific gravity :- "))
y = float(input("Give the interval for pressure :-"))


Tg = (Tg*(9/5)) + 491.67

m1 = s*28.96

tc = 168 + 325*s - 12.5*((s)**2)

pc = 677 + 15.0*s -37.5*((s)**2)

print("*******************************************************************")

zl = []
pl = []
bgl = []
de1 = []
den = []
while pg > 0 :
  pg1 = pg*14.2233
  p = pg1/pc
  #p = round(p,2)

  T = Tg/tc
  #T = round(T,2)

  t = 1/T


  import math as m
  import pandas as pd

  X1 = -0.06125*p*t*(m.exp(-1.2*((1-t)**2)))
  X2 = (14.76*t - 9.76*(t**2) + 4.58*(t**3))
  X3 = (90.76*t - 242.2*(t**2) + 42.4*(t**3))
  X4 = (2.18 + 2.82*t)

  Y = 0.0125*p*t*(m.exp(-1.12*((1-t)**2)))

  
  Yn = 0
  
# detemining the z implementing newton-raphson method on HALL-YARBOROUGH  method
  while abs(Y - Yn) > 0.00000000000001 :

    Yn = Y

    f = X1 + ((Y+(Y**2)+(Y**3)+(Y**4))/((1-Y)**3)) - X2*(Y**2) + X3*(pow(Y,X4))
    f1 =((1 + 4*Y + 4*(Y**2) - 4*(Y**3) + Y**4 )/((1 - Y)**4)) - 2*X2*Y + X4*X3*(pow(Y,X4-1))

    Y = Y - (f/f1)



  z = ((0.06125*p*t)/Y)*(m.exp(-1.2*((1-t)**2)))

  
# calculation gas formation volme factor    
  Bg = 0.0287*z*Tg/pg1

 
# calculation of viscosity and density    
  
  k = (9.4 + 0.02*(m1))*(pow(Tg,1.5))/(209+19*(m1)+Tg)
  V = 3.5 + (986/Tg) + 0.01*m1
  U = 2.4 - 0.2*V
  d = (0.0014935*pg1*m1)/(z*Tg)
  l = (pow(d,U))
  b = V*l
  de = 0.0001*k*(pow(2.71828,b))  
  

  den.append(round(d,4))
  zl.append(round(z,4))
  pl.append((pg))  
  bgl.append(round(Bg,4))
  de1.append(round(de,4))
  
# reduces the pressure by given interval
  pg = pg - y

#making the data frame

df = pd.DataFrame(pl,columns =['P(KG/CM2)'])
df['Z'] = zl
df['Bg'] = bgl
df['amug(viscosity)'] = de1
df['deng(density)'] = den

# to dispaly the full dataframe

pd.options.display.max_rows
pd.set_option('display.max_rows', None)
print(df)


```
